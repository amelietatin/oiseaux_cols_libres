### 4) Table: 'Totaux jours mois'

#### A) Dir_1: Direction Nord-Sud

```python
# change working directory
os.chdir(os.path.join(os.getcwd(), '../../totaux_jours_mois/dir_1'))
```

```python
files = [f for f in os.listdir() if os.path.isfile(f)]
files
```

```python
# Initialize an empty DataFrame to store the concatenated results
totaux_jours_mois_dir_1 = pd.DataFrame()

# Iterate over each file in the list 'files'
for file_i in files:

    # Print the current file name for debugging purposes
    print(file_i)

    # Read the CSV file into a DataFrame
    file = pd.read_csv(file_i)

    # Transpose the DataFrame to switch rows and columns
    file_transposed = file.transpose()

    # Set the first row as the header for the DataFrame
    file_transposed.columns = file_transposed.iloc[0]

    # Remove the first row, which is now used as the header
    file_transposed = file_transposed[1:]

    # Reset the index of the DataFrame and include the old index as a column
    file_transposed.reset_index(inplace=True)

    # Rename columns to 'Date' and 'Heure' for clarity
    file_transposed.rename(columns={file_transposed.columns[0]: "Date", file_transposed.columns[1]: "Heure"}, inplace=True)

    # Concatenate the transposed DataFrame with the accumulated DataFrame
    totaux_jours_mois_dir_1 = pd.concat([totaux_jours_mois_dir_1, file_transposed], ignore_index=True)
```
