## 3) Keep relevant columns for analysis

#### A) Data day for analysis

```python
# Convert columns to datetime
resume_par_jour.loc[:, 'stop_dt'] = pd.to_datetime(resume_par_jour['stop'])
resume_par_jour.loc[:, 'start_dt'] = pd.to_datetime(resume_par_jour['start'])
resume_par_jour.loc[:, 'date_dt'] = pd.to_datetime(resume_par_jour['date'])
```

```python
# Calculate the difference in hours
resume_par_jour.loc[:, 'day_duration'] = (resume_par_jour['stop_dt'] - resume_par_jour['start_dt']).dt.total_seconds() / 3600
```

```python
resume_par_jour.loc[:, 'stop_time'] = resume_par_jour['stop_dt'].dt.time
resume_par_jour.loc[:, 'start_time'] = resume_par_jour['start_dt'].dt.time

resume_par_jour.loc[:, 'year'] = resume_par_jour['date_dt'].dt.year
resume_par_jour.loc[:, 'month'] = resume_par_jour['date_dt'].dt.month
resume_par_jour.loc[:, 'day'] = resume_par_jour['date_dt'].dt.day
```

```python
resume_par_jour.columns
```

```python
data_day = resume_par_jour[['date_dt', 'year', 'month', 'day', 'start_time', 'stop_time', 'day_duration', 'observers', 'observersactive', 'observerspresent', 'species', 'direction1', 'direction2', 'maxlocal', 'weather', 'windspeed_bfr', 'wind_ms',
       'winddirection', 'cloudcover', 'cloudheight', 'precipitation',
       'perc_duration', 'visibility', 'temperature']]
```
